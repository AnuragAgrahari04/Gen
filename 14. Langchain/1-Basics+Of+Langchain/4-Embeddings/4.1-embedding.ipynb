#### Embedding Techniques

Converting text into vectors


```python
import os
from dotenv import load_dotenv
load_dotenv()  #load all the environment variables
```

```python
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings=OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text = "This is a tutorial on OPENAI embedding"
try:
    query_result = embeddings.embed_query(text)
except Exception as e:
    # Most commonly: OpenAI quota/billing issues (HTTP 429 insufficient_quota).
    msg = str(e)
    if "insufficient_quota" in msg or "Error code: 429" in msg or "RateLimitError" in msg:
        print("OpenAI embeddings request failed (quota/rate-limit). Falling back to HuggingFace embeddings...")
        from langchain_huggingface import HuggingFaceEmbeddings
        fallback_embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
        query_result = fallback_embeddings.embed_query(text)
    else:
        raise
query_result
```

```python
len(query_result)
```

```python
embeddings_1024=OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)
```

```python
embeddings_1024
```

```python
text = "This is a tutorial on OPENAI embedding"
try:
    query_result = embeddings_1024.embed_query(text)
except Exception as e:
    msg = str(e)
    if "insufficient_quota" in msg or "Error code: 429" in msg or "RateLimitError" in msg:
        print("OpenAI embeddings request failed (quota/rate-limit). Falling back to HuggingFace embeddings...")
        from langchain_huggingface import HuggingFaceEmbeddings
        # Rebind embeddings_1024 so later cells (vector DB, search) also work without OpenAI calls.
        embeddings_1024 = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
        query_result = embeddings_1024.embed_query(text)
    else:
        raise
len(query_result)
```

```python
query_result
```

```python
from langchain_community.document_loaders import TextLoader

loader=TextLoader('speech.txt')
docs=loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_documents=text_splitter.split_documents(docs)
final_documents
```

```python
## Vector Embedding And Vector StoreDB
from langchain_community.vectorstores import Chroma

db=Chroma.from_documents(final_documents,embeddings_1024)
db

```

```python
### Retrieve the results from query vectorstore db
query="It will be all the easier for us to conduct ourselves as belligerents"
retrieved_results=db.similarity_search(query)
print(retrieved_results)
```
